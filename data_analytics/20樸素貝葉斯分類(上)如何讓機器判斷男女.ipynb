貝葉斯原理
====
1. 為解決和『正向概率』相反的『逆向概率』，事先不知道袋子裏黑白球比例，通過摸出來的球顏色判斷袋子裡面黑白球的比例。建立在主觀判斷，在不了解客觀事實的情況下，同樣可先估計一個值，後根據實際結果不斷進行修正。
2. ex: 假設某病發病率是萬分之一，現有一方法檢測是否得病，準確率為99.9%，誤報率是0.1%。現在問題是，如果一個人被查出來患有該病，實際上患有的可能性是多大？
3. 『查出該病』準確率99.9%，但『實際患病』概率並非99.9%。例如：10000人尚存0.1%誤差，即10人沒病卻被診斷有病；而10000人也確實99.9%檢查出1人實際患病，患者是11人中的一員，即『實際患病』比例為1/11。

## 貝葉斯原理的概念：
### 1. 先驗概率：
通過『經驗判斷』事情發生概率，比如說『該病發病率是萬分之一』。比如南方梅雨季是6-7月，就是通過往年氣候總結該時段下雨的概率比其他時間高出很多。
### 2. 後驗概率：
發生結果後，推測原因的概率。比如查出某人患病，患病的原因可能是A, B, C。是原因Ａ的概率就是後驗概率。屬於『條件概率』的一種。
### 3. 似然函數(likelihood function)：
概率模型的訓練過程可理解為『求參數估計』的過程。例如：一個硬幣丟10次正面均朝上，該硬幣均勻的可能性是多少？『均勻就是參數』，似然函數即用來『衡量模型的參數』，關於『統計參數』的函數。

### 貝葉斯原理：
1. 實際上就是求解『後驗概率』，假設Ａ表示事件『測為陽性』，B1表示『患有該病』，B2表示『沒有該病』，可得以下結論：
    * 『患有該病』情況下，測出『陽性』的概率為P(A|B1)=99.9%；沒有該病，但測出『陽性』的概率為P(A|B2)=0.1%。
    * 『患有該病』的機率為P(B1)=0.01%，『沒有患病』的機率為P(B2)=99.9%。
    * 『檢驗為陽』且『實際患病』的概率P(B1,A)=P(B1)\*P(A|B1)=0.01%\*99.9%=0.00999%。
    * 此處P(B2,A)代表『聯合概率』，同樣可求得P(B2,A)=P(B2)\*P(A|B2)=99.99%\*0.1%=0.09999%。

樸素貝葉斯
====
重點要講此算法，**一種簡單但極為強大的預測建模算法**。假設每個輸入『變量是獨立的』，此為強硬的假設，實際情況並不一定，但這項技術對於絕大部分的複雜問題仍非常有效。
## 由兩種類型的概率組成：
1. 每個類別的概率P(Cj)：例如：假設我有7個棋子，其中3個白色、4個黑色，則黑色的概率為4/7，白色概率為3/7，即『類別概率』。
2. 每個屬性的條件概率P(Ai|Cj)：例如：將七個棋子放到兩個盒子裏，其中『A盒』裡面有2個白棋，2個黑棋；『B盒』有1個白棋，2個黑棋。在『A盒』抓到白棋的概率為1/2，抓到黑棋的概率為1/2，即『條件概率』，在『某個條件下(例如A盒中)發生的概率』。

### 樸素貝葉斯統計的是『屬性的條件概率』，即假設取出來的是白色的棋子，則『屬於盒子A的概率』為2/3。
1. 給出訓練數據
2. 數據對應的『分類』
3. 分別計算出『類別概率』和『條件概率』
4. 使用貝葉斯原理對新數據進行預測。

貝葉斯原理、貝葉斯分類、樸素貝葉斯 三者區別
====
1. 『貝葉斯原理』解決『逆向機率』，依此基礎設計『貝葉斯分類器』，『樸素貝葉斯分類』是貝葉斯分類器中的一種，最簡單、常用的。
2. 樸素貝葉斯之樸素是因為『假設屬性是互相獨立』，因此對實際情況有所約束，如果屬性間存在關聯，『分類準確率會降低』。不過在對於大部分情況，樸素貝葉斯分類效果不錯。

離散數據案例
====
分為兩種
1. 離散數據：不連續、有明顯邊界，比如整數1,2,3。
2. 連續數據：1至3之間的任何數，就是連續數據。
以下數據為訓練數據，再給你新的數據：身高、體重、鞋碼，判斷該人是男還是女？

|編號|身高|體重|鞋碼|性別|
|----|----|----|----|----|
|1|高|重|大|男|
|2|高|重|大|女|
|3|中|中|大|男|
|4|中|中|中|男|
|5|矮|輕|小|女|
|6|矮|輕|小|女|
|7|矮|中|中|女|
|8|中|中|中|中|

3. 針對該問題，首先確定一共有三個屬性：
    * 假設我們用Ａ代表屬性，用A1,A2,A3分別為身高=高、體重=中、鞋碼=中。
    * 兩個類別，假設用C代表類別，那麼C1,C2分別是: 男、女，在未知的情況下我們用Cj表示。
4. 想求A1,A2,A3屬性下，Cj的概率，用條件概率表示就是P(Cj|A1A2A3)。
<img src="./images/20-07.png">

5. 一共2個類別，只需要求得P(C1|A1A2A3)和P(C2|A1A2A3)的機率即可，然後比較哪個分類的可能性大，即是『分類結果』。P(A1A2A3)是固定的，尋求使得P(Cj|A1A2A3)的最大值，等價於求P(A1A2A3|Cj)\*P(Cj)最大值。

### 假定Ai之間相互獨立：
P(A1A2A3|Cj)=P(A1|Cj)P(A2|Cj)P(A3|Cj)
### 需要從Ai和Cj中計算出P(Ai|Cj)的概率：
帶入到上面的公式得出P(A1A2A3|Cj)，最後找到使得P(A1A2A3|Cj)最大的類別Cj。
### 分別求下這些條件下的概率：
P(A1|C1)=1/2, P(A2|C1)=1/2, P(A3|C1)=1/4, P(A1|C2)=0, P(A2|C2)=1/2, P(A3|C2)=1/2，所以**P(A1A2A3|C1)=1/16, P(A1A2A3|C2)=0**。

因為P(A1A2A3|C1)P(C1)>P(A1A2A3|C2)P(C2)，所以應該是C1類別，即男性。

## 如果給你一個新數據，身高180、體重120、鞋碼41，請問該人是男是女？
公式相同，困難在『連續變量不能採用離散變量的方法計算概率』。由於樣本太少，也無法分成區間計算。

## 假設男、女身高、體重、鞋碼都是『正態分佈』，算出某點的密度函數的值。
例如：男性身高均值為179.5、標準差為3.697的正態分佈。所以男性的身高為180的概率為0.1069。可使用excel的NormDist(x, mean, standard_dev,cumulative)函數，一共四個參數：
1. X：常態分配中，需要計算的數值。
2. Mean: 常態分配的平均值
3. Standard_dev: 常態分配的標準差
4. Cumulative: 取值為boolean值，決定函數形式。True為『累積分佈』；False為『機率密佈』。

### 這裡使用NormDist(180,179.5,3.697,0)=0.1069。
同理可計算男性體重為120的概率為0.000382324，男性鞋碼為41號的概率為0.120304111。

### 計算得出：
p(A1A2A3|C1)=P(A1|C1)P(A2|C1)P(A3|C1)=0.10690.0003823240.120304111=4.9169e-6

### 同理我们也可以计算出来该人为女的可能性：
P(A1A2A3|C2)=P(A1|C2)P(A2|C2)P(A3|C2)=0.000001474890.0153541440.120306074=2.7244e-9

很明显这组数据『分类为男的概率大于分类为女的概率』。

樸素貝葉斯分類器工作流程
====
常用於『文本分類』、垃圾文本過濾、情感預測、推薦系統等。
<img src="./images/20-08.jpg">

## 第一階段：準備階段
在这个阶段我们需要确定特征属性，比如上面案例中的“身高”、“体重”、“鞋码”等，并对每个特征属性进行适当划分，然后由人工对一部分数据进行分类，形成训练样本。

这一阶段是整个朴素贝叶斯分类中唯一需要人工完成的阶段，其质量对整个过程将有重要影响，分类器的质量很大程度上由特征属性、特征属性划分及训练样本质量决定。

## 第二階段：訓練階段
这个阶段就是生成分类器，主要工作是计算每个类别在训练样本中的出现频率及每个特征属性划分对每个类别的条件概率。

输入是特征属性和训练样本，输出是分类器。

## 第三階段：應用階段
这个阶段是使用分类器对新数据进行分类。输入是分类器和新数据，输出是新数据的分类结果。
