```python
import jieba
import os
from sklearn.feature_extraction.text import TfidfVectorizer
# 多项式贝叶斯分类器
from sklearn.naive_bayes import MultinomialNB  
# 驗證準確率
from sklearn import metrics
```

```python
def cut_word(file_dir, label):
    file_list = os.listdir(file_dir)
    res_txt = []     #訓練文件與訓練label個數要成對。
    res_label = []
    for txt in file_list:
        txt_dir = file_dir + '/' + txt
        words_with_space = ''
        with open(txt_dir, 'r', encoding='gb18030') as f:
            text_cut = jieba.cut(f.read())
            
            for word in text_cut:
                words_with_space += word + ' '

            res_txt.append(words_with_space)
            res_label.append(label)
            
    return res_txt, res_label
# 訓練集
train_word_list1, train_label_list1 = cut_word('text classification/train/女性', '女性')
train_word_list2, train_label_list2 = cut_word('text classification/train/体育', '体育')
train_word_list3, train_label_list3 = cut_word('text classification/train/文学', '文学')
train_word_list4, train_label_list4 = cut_word('text classification/train/校园', '校园')

train_word = train_word_list1 + train_word_list2 + train_word_list3 + train_word_list4
train_label = train_label_list1 + train_label_list2 + train_label_list3 + train_label_list4
```

```python
# 測試集
test_word_list1, test_label_list1 = cut_word('text classification/test/女性', '女性')
test_word_list2, test_label_list2 = cut_word('text classification/test/体育', '体育')
test_word_list3, test_label_list3 = cut_word('text classification/test/文学', '文学')
test_word_list4, test_label_list4 = cut_word('text classification/test/校园', '校园')

test_word = test_word_list1 + test_word_list2 + test_word_list3 + test_word_list4
test_label = test_label_list1 + test_label_list2 + test_label_list3 + test_label_list4
```

```python
stop_words = open('text classification/stop/stopword.txt', 'r', encoding='utf-8').read()
stop_words = stop_words.encode('utf-8').decode('utf-8-sig') # 列表头部\ufeff处理
stop_words = stop_words.split('\n') # 根据分隔符分隔
stop_words
```

```python
tf = TfidfVectorizer(stop_words=stop_words, max_df=0.5) #計算單詞權重和stop_words
```

```python
train_features = tf.fit_transform(train_word)
# 上面fit过了，这里transform
test_features = tf.transform(test_word) 
```

```python
clf = MultinomialNB(alpha=0.001).fit(train_features, train_label)
predicted_label=clf.predict(test_features)

# 计算准确率
print('准确率为：', metrics.accuracy_score(test_label, predicted_label))
```
