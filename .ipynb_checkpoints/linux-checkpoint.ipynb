2. 進入mongodb
```shell
# 進入
mongo
# 選擇db, 若db不存在，會新增並進入
use database
```
3. 為database新增使用者
```shell
# 新增，以前是addUser，但現在已改為createUser
db.createUser({user:'yuyu',pwd:'1234',roles:['readWrite','dbAdmin']})
# 登入
db.auth('account', 'password')
```
4. 創建第一筆資料
由於 MongoDB 是 Schemaless 的資料庫，我們不需要向 SQL 一樣定義 Table 資料表 的長相。因此創建資料是一件非常容易的事情！
只需要輸入簡單的兩行指令即可：

```shell
use linedb
# 在名為myNewCollection1的collection中創建一筆document。
# 名為hello的field(欄位)中有值mongoDB(字串str)
db.myNewCollection1.insertOne( { hello: 'mongoDB' } )
```
